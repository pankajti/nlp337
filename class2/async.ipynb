```python
from nltk.corpus import gutenberg 
```

```python
gutenberg.fileids()
```

```python
for fileid in gutenberg.fileids():
    num_chars = len(gutenberg.raw(fileid))
    num_words = len(gutenberg.words(fileid))
    num_sents = len(gutenberg.sents(fileid))
    num_vocab = len(set([w.lower() for w in gutenberg.words(fileid)]))
    print(int(num_chars/num_words), int(num_words/num_sents), int(num_words/num_vocab),fileid)
```

```python
from nltk.corpus import brown
brown.categories()
```

```python
for fileid in brown.fileids():
    print(fileid)
```

```python
news_text = brown.words(categories ='news')
```

```python
fdist = nltk.FreqDist([w.lower() for w in news_text])
```

```python
fdist['may']
```

```python
modals = ['may' , 'might' , 'can' , 'could', 'must' , 'will']

for m in modals :
    print(m + ":  "  + str(fdist[m]))
```

```python
cfd = nltk.ConditionalFreqDist((genre, word) for genre in brown.categories()
                              for word in brown.words(categories =genre))
```

```python
genres = ['news', 'religion', 'hobbies', 'science_fiction', 'romance', 'humor']
modals = ['can', 'could', 'may', 'might', 'must', 'will']


cfd.tabulate(conditions = genres , samples = modals)
```

```python
from nltk.corpus import udhr
```

```python
languages = ['Chickasaw', 'English', 'German_Deutsch', 'Greenlandic_Inuktikut', 'Hungarian_Magyar', 'Ibibio_Efik']
```

```python
cfd = nltk.ConditionalFreqDist((language, len(word)) for language in languages 
                              for word in udhr.words(language+'-Latin1'))
```

```python
import matplotlib.pyplot as plt
cfd.plot(cumulative = True)
```

```python
from nltk.corpus import inaugural
cfd = nltk.ConditionalFreqDist(
 (target, fileid[:4])
 for fileid in inaugural.fileids()
 for w in inaugural.words(fileid)
 for target in ['america', 'citizen']
 if w.lower().startswith(target))
```

```python
cfd.plot(cumulative = True)
```
